## Background and Prior Work

The complex relationship between exercise and sleep has been a topic of scientific interest for several decades, with a growing body of evidence suggesting that the type, duration, and timing of physical activity can significantly impact sleep quality and quantity
In the past, the central nervous system (CNS) was researched in order to identify sleep patterns in the form of electroencephalogram (EEG) waveforms <sup><a href="#footnote-1">[1]</a></sup> . When researchers observed these EEG waves during sleep after daytime exercise in the past, they found little to no differences. However, more recently, researchers started to observe both CNS physiology as well as somatic physiology, which led to the conclusion that the regulation of sleep is not just based on the CNS, but is rather affected by different aspects of the body. 

A survey on a blog post by the Sleep Foundation also researched on the effects of exercise on improving sleep quality <sup><a href="#footnote-2">[2]</a></sup>, with special attention to what types or times of exercises led to better results. They found that moderate to extreme levels of exercise had the most noticeable effect on increasing sleep quality, due to how the subjects took less time to fall asleep as well as it decreased the amount of time individuals were awake throughout the night. This is due to how physical daytime activity affects the endocrine, autonomic nervous system (ANS) and somatic functions, which can also be measured during sleep cycles. In relation to EEG, through a research study in 1999, researchers were able to use computer analyses of EEG during sleep combined with neuroimaging techniques to result in more depth in their CNS function analyses during sleep (Feinberg et al., 1978; Kajimura et al., 1999; Uchida et al., 1999). Consistent exercise was shown to have the largest increase in sleep quality, especially when it was consistently at the same time of day.

Explanation of previous work:

Kubitz et al. (1996) and Youngstedt et al. (1997) meta analytic studies examined both acute and chronic influences that exercise had on sleep parameters. These studies concluded that daytime exercise a bit of time before sleep (subsequent sleep) resulted in longer NREM sleep stages and shorter periods of stage REM, however, acute exercise effects were inconsistent while the chronic exercise effects were more consistent. Some variables that they found had an impact on the data were: “fitness of subjects, exercise heat load, duration and time of day, and subjects’ light exposure and sleep schedule” (Youngstedt et al. (1997)). They found that since daytime exercise is known to influence body composition, metabolic rate of the person, cardiac function, glycemic control, immune function, and mood state positively, it can be a factor in either improving or disrupting sleep. 

The revision of these studies led to a conclusion that daytime exercise—acute effects, acute with chronic effects, chronic effects—influences a person’s quality of sleep differently. Furthermore, physical activity has the potential to serve as a strong trigger for the self-regulating process that influences all of the body's physiological systems during sleep.

#### Data analysis: T-test

We would like to do analysis between Physical Activity Level and sleep metrics (like Sleep Duration or Quality of Sleep). Thus it is appropriate to use T-test to evaluate the impact of exercise on sleep.

To simplify our setting, we can categorize the Physical Activity Level into two groups: high and low activity. The categorization can be based on a median split or a specific threshold. We will then perform a T-test to compare the Sleep Duration and Quality of Sleep between these two groups.


```python
from scipy.stats import ttest_ind

# Categorize Physical Activity Level into high and low based on the median
median_physical_activity = health_df['Physical Activity Level'].median()
high_activity_group = health_df[health_df['Physical Activity Level'] >= median_physical_activity]
low_activity_group = health_df[health_df['Physical Activity Level'] < median_physical_activity]

# T-test for Sleep Duration
t_test_sleep_duration = ttest_ind(high_activity_group['Sleep Duration'], low_activity_group['Sleep Duration'])

# T-test for Quality of Sleep
t_test_quality_of_sleep = ttest_ind(high_activity_group['Quality of Sleep'], low_activity_group['Quality of Sleep'])

t_test_sleep_duration, t_test_quality_of_sleep
```

Interpretation

**Sleep Duration**

- The T-test result for sleep duration between high and low physical activity groups shows a statistically significant difference (p-value much lower than 0.05).

- This suggests that there is a significant difference in sleep duration between individuals with high and low levels of physical activity.

**Quality of Sleep**

- The T-test for quality of sleep also indicates a statistically significant difference between the two groups (p-value much lower than 0.05).

- This implies that the quality of sleep is significantly different between individuals with high and low physical activity levels.

**Conclusion we can make**

These findings support the hypothesis that physical activity level (high vs. low) has a significant impact on both the duration and quality of sleep. Individuals with higher physical activity levels tend to have longer and better-quality sleep compared to those with lower physical activity levels. This aligns with the positive relationships we observed in the earlier correlation and scatter plot analyses.

## Predictive Analysis 
### Data analysis: Linear Regression

Linear regression is used to understand the linear relationship between a dependent variable and one or more independent variables. In this case we are interested in understanding how different factors (like exercise frequency, age) affect sleep metrics (like sleep efficiency, deep sleep percentage). Linear regression can provide valuable insights into how exercise frequency and other variables are related to sleep quality and duration.

```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

# Selecting the dependent and independent variables
# Using Sleep efficiency as the dependent variable
dependent_variable = 'Sleep efficiency'

# Assuming we are interested in Age and Exercise frequency as independent variables
independent_variables = ['Age', 'Exercise frequency']

# Preparing the data
X = cleaned_efficiency_df[independent_variables]
y = cleaned_efficiency_df[dependent_variable]

# Adding a constant to the model (intercept)
X = sm.add_constant(X)

# Building the linear regression model
model = sm.OLS(y, X).fit()

# Getting the model summary
model_summary = model.summary()
model_summary
```

Linear Regression Analysis Results:

Model Summary

- **Dependent Variable:** Sleep efficiency
- **Independent Variables:** Age, Exercise frequency
- **R-squared:** 0.064 - This value indicates that approximately 6.4% of the variability in sleep efficiency is explained by the model. It's relatively low, suggesting that other factors not included in the model might also be significant predictors of sleep efficiency.

Coefficients

- **Constant (Intercept):** 0.7417 - The baseline value for sleep efficiency when all independent variables are zero.
- **Age:** Coefficient of 0.0004, with a p-value of 0.608. This suggests that age is not a statistically significant predictor of sleep efficiency in this model.
- **Exercise Frequency:** Coefficient of 0.0232, with a p-value of 0.000. This indicates that exercise frequency is a significant predictor of sleep efficiency. For each unit increase in exercise frequency, sleep efficiency increases by 0.0232, holding other factors constant.

Next we will visualize the results by plotting the relationship between exercise frequency and sleep efficiency along with the regression line

```python
# Plotting the relationship between Exercise Frequency and Sleep Efficiency
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Exercise frequency', y='Sleep efficiency', data=efficiency_df, alpha=0.6)

# Calculating the regression line
reg_line = model.params['const'] + model.params['Exercise frequency'] * efficiency_df['Exercise frequency']
plt.plot(efficiency_df['Exercise frequency'], reg_line, color='red', linewidth=2)

plt.title('Sleep Efficiency vs. Exercise Frequency with Regression Line')
plt.xlabel('Exercise Frequency')
plt.ylabel('Sleep Efficiency')
plt.show()
```

Analysis and Conclusion
- **Significance of Exercise Frequency:** The linear regression analysis and the visualization both support the idea that exercise frequency has a positive impact on sleep efficiency. This is in line with hypothesis and indicates that engaging in more frequent exercise might be associated with better sleep efficiency.

- **Model Limitations:** The R-squared value is relatively low, indicating that other variables not included in the model may also play a significant role in determining sleep efficiency. These could include factors like diet, stress levels, or the type and timing of exercise.



## Gradient Boosting Machines (GBMs)
Gradient Boosting Machines (GBMs) are a type of ensemble learning technique that build models in a stage-wise fashion. The core idea is to combine several weak models (usually decision trees) to create a strong predictive model. GBMs are particularly effective because they correct the mistakes of previous models in each iteration, making the model increasingly accurate.

For our analysis, GBMs are chosen because:
- **Ability to Handle Complex Relationships:** They can capture non-linear and complex relationships between variables.
- **Robust to Overfitting:** GBMs are less prone to overfitting, especially with careful tuning of parameters.
- **Feature Importance:** They provide insights into which predictors are most influential, helping us understand the data better.

We will apply Gradient Boosting Machines to the 'Sleep Health and Lifestyle' dataset

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Splitting the dataset
X = cleaned_health_df[['Physical Activity Level', 'Daily Steps']]
# X = cleaned_health_df[['Physical Activity Level', 'Stress Level', 'BMI Category', 'Heart Rate', 'Daily Steps']]

y = cleaned_health_df['Sleep Duration']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)

# Creating and training the GBM model
gbm_pipeline = Pipeline([
    ('gradient_boosting', GradientBoostingRegressor(random_state=42))
])
gbm_pipeline.fit(X_train, y_train)

# Predicting and evaluating the model
y_pred = gbm_pipeline.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error: {mse}") 
print(f"R2 Score: {r2}")
```

Analysis:
- These results indicate a moderate level of predictive accuracy. The R² score of 0.611 means that around 61% of the variance in sleep duration is explained by physical activity level and daily steps.

Now, let's proceed to build and evaluate the second model, where we add 'Stress Level' to the set of predictor variables.

```python
# Adding 'Stress Level' to the features for the second model
features_2 = ['Physical Activity Level', 'Stress Level', 'Daily Steps']

# Splitting the dataset with the updated features
X_2 = cleaned_health_df[features_2]
X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(X_2, y, test_size=0.5, random_state=42)

# Creating and training the updated GBM model
gbm_model_2 = GradientBoostingRegressor(random_state=42)
gbm_model_2.fit(X_train_2, y_train_2)

# Predicting and evaluating the updated model
y_pred_2 = gbm_model_2.predict(X_test_2)
mse_2 = mean_squared_error(y_test_2, y_pred_2)
r2_2 = r2_score(y_test_2, y_pred_2)

print(f"Mean Squared Error: {mse_2}") 
print(f"R2 Score: {r2_2}")
```

Analysis:
- The addition of 'Stress Level' significantly improved the model's performance. The R² score increased to 97.8%, indicating that the model now explains a much larger portion of the variance in sleep duration. The MSE also decreased substantially, suggesting more accurate predictions.
